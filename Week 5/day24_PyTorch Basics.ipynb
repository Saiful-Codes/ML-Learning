Learnable Tensors

```python
x = torch.tensor(2.0, requires_grad=True)
x
```

Autograd

```python
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2
y
```

```python
y.backward() #backpropagation
```

```python
x.grad #Inspect the gradient
```

Example - Linear Model Gradient

```python
x = torch.tensor(4.0)
w = torch.tensor(2.0, requires_grad=True)

y = w * x
y.backward()

w.grad
```
